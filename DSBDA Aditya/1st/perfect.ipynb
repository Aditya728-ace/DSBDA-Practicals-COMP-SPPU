```python
#### FROM HERE THE QUESTIONS STARTS

# Calculate the mean value for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avg_stroke)

# Replace NaN by mean value in "stroke" column
df["stroke"] = df["stroke"].replace(np.nan, avg_stroke)
```

```python
#Calculate the mean value for the 'horsepower' column
avg_hp = df["horsepower"].astype("float").mean(axis = 0)
print("Average of HP:", avg_hp)
```

```python
# check null values
print(df["peak-rpm"].isna().sum())  # Should show 2 before replacement

```

```python
## fill null values by avg/mean values

avg_rpm = df["peak-rpm"].mean()

# Replace NaN values with the mean
df["peak-rpm"] = df["peak-rpm"].fillna(avg_rpm)
```

```python
print(df["peak-rpm"].isna().sum())  # Should show 0 after replacement

```

```python
######Indicator variable (or dummy variable)     Turns categorical values to quantitative

An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.

Why we use indicator variables?

So we can use categorical variables for regression analysis in the later modules.

Example:

We see the column "fuel-type" has two unique values, "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.

We will use the panda's method 'get_dummies' to assign numerical values to different categories of fuel type.
```

```python
##Binning Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
# IT IS SLOTTING TYPE  LIKE RANGING
```

```python
# create evenly spaced bins OR SLOTS

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']   # Naming the labels
```

```python
#####BINNING CODE

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(5)
```

```python
df["horsepower-binned"].value_counts()   #counts according to slots
```
